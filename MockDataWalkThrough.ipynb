```python
# can augment data frame (changes will stick in memory) -- changing name can overwrite original data frame
# passing as dictionary for column names (original column name : new column name)
mock_data = mock_data.rename(columns={"first_name":"fname","last_name":"lname"})
```

```python
# see first few rows with new column name changes
mock_data.head()
```

```python
# gives all values from one column (fnames)
mock_data['fname']
```

```python
# calling object out of data frame (same way to do above)
mock_data.fname
```

```python
# calling two objects out of data frame (requires two square braces)
mock_data[['fname','lname']]
```

```python
# plot of data frame (count certain values)
pd.value_counts(mock_data['gender']).plot.bar()
```

```python
# group by data frame
mock_data.groupby(['gender'])['age'].mean()
```
